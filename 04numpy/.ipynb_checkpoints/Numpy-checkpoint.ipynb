```python
s = numpy.linspace(-10, 10, num=100)
x, y = numpy.meshgrid(s, s)
```

```python
z = x**2 + y **2
```

```python
pyplot.imshow(z)
pyplot.colorbar()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = pyplot.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y,  z, rstride=1, cstride=1, cmap='viridis')
```
